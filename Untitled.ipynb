# Esempio

  Sia $ y=e^{-\frac{x}{10}}sin(x) $. Considera un intervallo di x di 10000 nell'intervallo [0,10]

1. Grafico della funzione y vs. x nell'intervallo [0,10]

```python
N=10000
x=np.linspace(0,10,N+1)
y=np.exp(-x/10)*np.sin(x)
plt.plot(x,y)
```

2. Calcola la media e la deviazione standard di y per x appartenente a [4,7]

```python
y1=y[(x>=4)*(x<=7)]
med=np.mean(y1)
med

```

```python
dev=np.std(y1)
dev
```

3. Per x appartenente a [4,7] trovare $y_m$ tale che  l'80% dei valori di y è minore di $y_m$

```python
np.percentile(y1,80)
```

4. Plot dy/dx vs. x


```python
dydx=np.gradient(y,x)
plt.plot(x,y)
plt.plot(x,dydx)
```

5. Trovare gli zeri di dydx

```python
zeros=[]
for i in range(1,len(dydx),1):
    if(dydx[i]*dydx[i-1]<0):
         zeros.append((x[i]+x[i-1])/2)

y_zero=np.zeros(len(zeros))
plt.plot(x,dydx)
plt.plot(zeros,y_zero,'*g')

       
```

# Esempio 3

Considera la curva $r(\theta)=1+\frac{3}{4}sin(3\theta)$ per $0<=\theta<=2\pi $

1. Fare un grafico della curva


```python
theta=np.linspace(0,2*np.pi,1000)
r=1+3/4*np.sin(3*theta)
x=r*np.cos(theta)
y=r*np.sin(theta)
plt.plot(x,y)
```

 2. Calcolare l'area usando la formula $ A=\[\int_{0}^{2*\pi} \frac{1{2}r^2 \,d\theta\]$
 

```python
A=1/2*sum(r**2)*(theta[1]-theta[0])
A
```

```python
kt = np.linspace(0, 3, 100)
P = (1/(1+np.exp(-kt)))**4
E = np.cumsum(P) * (kt[1]-kt[0])

plt.plot(kt,E)
plt.xlabel('$kt$', fontsize=20)
plt.ylabel(r'$\left(\frac{k}{A \sigma \epsilon T_0^4}\right) E(kt)$', fontsize=20)
plt.show()

```
